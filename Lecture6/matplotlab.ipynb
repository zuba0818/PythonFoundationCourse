```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
plt.plot([1,2,3],
         [4,5,1])

plt.show()
```

```python
x = np.linspace(0,10,1000)

plt.plot(x, x+0, linestyle='solid') # or -
plt.plot(x, x+1, linestyle='dashed') # or --
plt.plot(x, x+2, linestyle='dashdot') # or -.
plt.plot(x, x+3, linestyle='dotted') # or :
```
